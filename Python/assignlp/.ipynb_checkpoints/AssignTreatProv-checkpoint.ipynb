This is an example of using these `assign_funcs` to assign people/sources to providers given XY coordinates. E.g. you may have estimates for people who need treatment per zip-code, and want to assign them to the nearest treatment provider.

The main motivation is to identify areas that are undercovered, as well as providers that are overburdened.

Note that to use the shadow/slack variable estimates in python pulp, you need to use a solver that returns them. Only ones I know of are `CBC` (which is open), and `CPLEX` (which needs a license, but is free for academics).

To install CBC on Windows, I have had luck doing:

    pip install cbcpy
    pip install pulp

So install cbcpy *before* pulp to get it to recognize it.

------

The linear programming model is, for source/people locations $i$ and provider locations $j$:

$$\text{Min}(D_{ij}\text{distance}_{ij})$$

Subject To:

  1. $\Sigma D_{is}= T_i \ \ \ \forall i$ ($T_i$ is the total people within a source)
  2. $\Sigma D_{pj} <= T_j \ \ \ \forall j$ ($T_j$ is the total capacity for a provider)

$D_{ij}$ are the assignments of people/sources to a particular provider. So 0 means no assignments, 10 means 10 people assigned from source $i$ to provider $j$, etc. And then $D_{ij} >= 0$ (but less than or equal to whatever $\text{max}(T_s)$ is). Technically you want integer solutions, but can leave the model continuous and if all inputs are integers it will result in integer solutions.

Constraint 1 is that everyone is covered. Constraint 2 is that providers do not go over capacity. The model functions automatically add in a slack provider at a particular set distance. So the model should be feasible. Also has functions to limit potential assignments to a particular distance threshold. So e.g. people will not be assigned a provider over 12 miles away (if that is your set threshold).

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pulp

# locations of people
id = range(10)
px = [1,1,3,3,4,5,6,6,9,9]
py = [2,8,1,8,4,6,1,2,5,8]
peop = pd.DataFrame(zip(id,px,py),columns=['id','px','py'])
peop['const'] = 1

# locations of 2 providers & capacity 3/5
hid = [1,2]
hx = [1,8]
hy = [1,8]
hc = [3,5] # so cant cover 2 people
prov = pd.DataFrame(zip(hid,hx,hy,hc),columns=['hid','hx','hy','hc'])

# graph of the data
fig, ax = plt.subplots()
ax.scatter(peop['px'], peop['py'],
           c='grey',edgecolor='k',s=30)
ax.scatter(prov['hx'], prov['hy'],
           c='red',edgecolor='k',marker='s',s=60)
plt.show()
```

```python
# load in model functions
from assign_funcs import ProvAssign

m1 = ProvAssign(peop,
                ['id','px','py','const'],
                prov,
                ['hid','hx','hy','hc'],
                12,
                9999)

m1.solve(pulp.PULP_CBC_CMD())
m1.assign # id is the source people, hid is the providers
```

```python
# if we up the provider capacity for
# prov by 1, the model objective goes 
# down by -9994
print(m1.shadow)


prov['hc'] = [4,5]

m2 = ProvAssign(peop,
                ['id','px','py','const'],
                prov,
                ['hid','hx','hy','hc'],
                12,
                9999)
m2.solve(pulp.PULP_CBC_CMD())
m1.obj - m2.obj # 9994!
```

```python
# Can make these multiple people, 100 total
peop['tot'] = [10,15,5,10,12,11,20,6,9,2]
prov['cap'] = [40,50] # should assign 10 people to slack

m3 = ProvAssign(peop,
                ['id','px','py','tot'],
                prov,
                ['hid','hx','hy','cap'],
                12,
                9999)
m3.solve(pulp.PULP_CBC_CMD())

# One source area can have multiple providers
# due to capacity limits source 4 goes
# to both provider 1 (5 people) & 2 (7 people)
m3.assign
```

```python
# Can add in hypothetical providers with
# no capacity (make new providers)

# Adding another provider with no capacity
p2 = prov.iloc[[0],:].copy()
p2['hid'] = 10
p2['hx'] = 5
p2['hy'] = 5
p2['hc'] = 0
p2['cap'] = 0
prov_add = pd.concat([prov,p2],axis=0)

# The 4th argument limits potential distances
# here limit to 10 makes it so source9
# and provider1 
m4 = ProvAssign(peop,
                ['id','px','py','tot'],
                prov_add,
                ['hid','hx','hy','cap'],
                10,
                9999)

# we now don't have source9 and prov1
# in potential assignment list, since
# it is over 10 distance
m4.cross
```

```python
m4.solve(pulp.PULP_CBC_CMD())

# can see the new hypothetical provider 10
# if we added capacity would
# decrease distance travelled by -9996.2426
m4.shadow
```

```python
# Now lets add capacity for new provider
# by 10, should the objective go down
# by 10*-9996.2426 ?
prov_add['cap'] = [40,50,10]

m5 = ProvAssign(peop,
                ['id','px','py','tot'],
                prov_add,
                ['hid','hx','hy','cap'],
                12,
                9999)
m5.solve(pulp.PULP_CBC_CMD())

# Not quite -9996.2 but close!
(m5.obj - m4.obj)/10
```

```python
# we could add capacity
# to provider 1 by 10
# as well

prov_add['cap'] = [50,50,0]

m6 = ProvAssign(peop,
                ['id','px','py','tot'],
                prov_add,
                ['hid','hx','hy','cap'],
                10,
                9999)
m6.solve(pulp.PULP_CBC_CMD())

# Not quite -9993.4 but close!
(m6.obj - m4.obj)/10
```

```python
# Can look at sources and see stats
# for distance travelled as well
# as potential non-coverage

# Here for m4 model, only source
# 1 has no coverage of 10 people

m4.source_stats
```

```python
# Can look at providers
# and see which ones would result in
# most reduced travel given increased coverage
# Want high NotCover/bisqw (bisquare weighted total) 
# and low shadow price
m4.prov_stats
```

```python
# Have another intermediate
# m4.dist_stats, can do groupbys
# for more intermediate numbers
# this shows total potential within
# 5 distance, only the new hid=10
# provider has any within that small
# of distance (others are 7 away)

m4.distprov(5)
```
